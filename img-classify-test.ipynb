```python
import cv2
import requests
import matplotlib.pyplot as plt
url = "http://localhost:5000/image_classify"
image_source = "animal_classify_img/test.jpg"
raw_image = cv2.imread(image_source)
image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
res = requests.post(url, json={'image': image.tolist()})
print(res.json())
```
